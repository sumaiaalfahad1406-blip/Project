```python
import pandas as pd
import matplotlib.pyplot as plt


# Load dataset
file_path = ""
df = pd.read_excel(file_path)

# Check required columns
required = ["subject", "relation", "object"]
for c in required:
    if c not in df.columns:
        raise ValueError(f"Column '{c}' not found. Found columns: {list(df.columns)}")

# Keep only complete triples for analysis 
df_clean = df.dropna(subset=required).copy()


# Data analysis numbers 
n_obs = df.shape[0]
n_features = df.shape[1]
n_obs_clean = df_clean.shape[0]

n_subjects = df_clean["subject"].nunique()
n_relations = df_clean["relation"].nunique()
n_objects = df_clean["object"].nunique()

# Triples per subject 
triples_per_subject = df_clean["subject"].value_counts()

# Object length stats 
obj_word_counts = df_clean["object"].astype(str).apply(lambda x: len(x.split()))

# Missing values
missing_counts = df.isnull().sum()

print("\n==============================")
print("3.5 DATA ANALYSIS (SUMMARY)")
print("==============================")
print(f"Observations (rows): {n_obs}")
print(f"Features (columns):  {n_features}")
print(f"Complete triples (no missing in subject/relation/object): {n_obs_clean}")

print("\n--- Unique counts ---")
print(f"Unique subjects:  {n_subjects}")
print(f"Unique relations: {n_relations}")
print(f"Unique objects:   {n_objects}")

print("\n--- Knowledge density (triples per subject) ---")
print("Mean:", round(triples_per_subject.mean(), 2))
print("Std:", round(triples_per_subject.std(), 2))
print("Min:", int(triples_per_subject.min()))
print("Q1:", round(triples_per_subject.quantile(0.25), 2))
print("Median:", round(triples_per_subject.median(), 2))
print("Q3:", round(triples_per_subject.quantile(0.75), 2))
print("Max:", int(triples_per_subject.max()))

print("\n--- Object text length (words) ---")
print("Mean:", round(obj_word_counts.mean(), 2))
print("Std:", round(obj_word_counts.std(), 2))
print("Min:", int(obj_word_counts.min()))
print("Q1:", round(obj_word_counts.quantile(0.25), 2))
print("Median:", round(obj_word_counts.median(), 2))
print("Q3:", round(obj_word_counts.quantile(0.75), 2))
print("Max:", int(obj_word_counts.max()))

print("\n--- Missing values (all columns) ---")
print(missing_counts)


# Tables like in paper 
# Table 1-style: Relation distribution 
rel_counts = df_clean["relation"].value_counts()
rel_table = pd.DataFrame({
    "Count": rel_counts,
    "Percent": (rel_counts / rel_counts.sum() * 100).round(2)
})
print("\n==============================")
print("Table 1: Distribution of relation types")
print("==============================")
print(rel_table.head(20))  # show top 20 (change if you want)

# Table 2-style: Evidence/knowledge per subject 
table2 = pd.DataFrame({
    "Avg": [triples_per_subject.mean()],
    "Std dev": [triples_per_subject.std()],
    "Min": [triples_per_subject.min()],
    "1st quartile": [triples_per_subject.quantile(0.25)],
    "Median": [triples_per_subject.median()],
    "3rd quartile": [triples_per_subject.quantile(0.75)],
    "Max": [triples_per_subject.max()]
}).round(2)
print("\n==============================")
print("Table 2: Statistics of triples per subject")
print("==============================")
print(table2.to_string(index=False))


# Figures (simple + understandable)

# Figure 1: Top relations bar chart
plt.figure()
df_clean["relation"].value_counts().head(10).plot(kind="bar")
plt.title("Figure 1: Top 10 Relation Types")
plt.xlabel("Relation")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

# Figure 2: Top subjects bar chart
plt.figure()
df_clean["subject"].value_counts().head(10).plot(kind="bar")
plt.title("Figure 2: Top 10 Subjects (Most Connected)")
plt.xlabel("Subject")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

# Figure 3: Triples per subject histogram
plt.figure()
plt.hist(triples_per_subject, bins=20)
plt.title("Figure 3: Distribution of Triples per Subject")
plt.xlabel("Triples per Subject")
plt.ylabel("Number of Subjects")
plt.tight_layout()
plt.show()

# Figure 4: Object length histogram (word count)
plt.figure()
plt.hist(obj_word_counts, bins=20)
plt.title("Figure 4: Distribution of Object Text Length (Words)")
plt.xlabel("Words per Object")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# Figure 5: Missing values bar chart
plt.figure()
missing_counts.plot(kind="bar")
plt.title("Figure 5: Missing Values per Column")
plt.xlabel("Column")
plt.ylabel("Missing Count")
plt.tight_layout()
plt.show()

# print one sample triple (like a "sample patient" but for KG)
print("\n==============================")
print("Sample triple (example row)")
print("==============================")
print(df_clean.sample(1).to_string(index=False))
```
